定義一個輸入年、月就能找出該月理論上的結算日日期的函數(每月的第三個星期三)

```python
#找結算日期
def find_third_wen(y, m):
    if datetime.date(y,m,1).weekday() < 2: #1號禮拜三還沒過
        d = 15 + (2 - datetime.date(y,m,1).weekday())
    elif datetime.date(y,m,1).weekday() == 2:#1號就是禮拜三
        d = 15
    else:
        d = (7 - datetime.date(y,m,1).weekday()) + 17#1號禮拜三過了
    date = datetime.datetime(y, m, d)
    return date
```

例如2022年8月的結算日就是8/17

```python
find_third_wen(2022, 8)
```

接著以這個函數找出所有理論上的結算日日期，並儲存在settlement_date串列

```python
year = [i for i in range(1998, 2023)]
month = [i for i in range(1, 13)]
#製作結算日清單
settlement_date = []
for y in year:
    for m in month:
        if y == 1998 and m in  [1, 2, 3, 4, 5, 6, 7, 8]:#1998年8月以前不在資料範圍內
            continue
        elif y == 2022 and m in [9, 10, 11, 12]:#2022年9月以後的結算日也不在資料範圍內
            continue
        else:
            settlement = find_third_wen(y, m)
            settlement = str(settlement).split(' ')[0]
            settlement_date.append(settlement)
```

因為我們只要考慮近月及次月，所以我們只需要抓出每個交易日的前兩筆資料(因為這個資料源已經依照到期月份排序了，所以前兩筆就是近月、次月的)

接下來我先抓出一個包含所有交易日的清單，方便最後依照這個清單整理出最後的時間序列

```python
trading_day = []
for i in range(len(day_data)):
    if not day_data.iloc[i]['日期'] in trading_day:
        trading_day.append(day_data.iloc[i]['日期'])
    else:
        continue
trading_day = pd.DataFrame(trading_day, columns= ['交易日'])
trading_day
```

再依據這個所有交易日的清單，從結算日回推三個交易日找出我們要轉倉的日期，同時也是檢驗我們這個理論結算日中有沒有例外機會(因為理論結算日會遇到休假順延)

```python
switch_date = []
for d in settlement_date:
    try:
        s = trading_day[trading_day['交易日'] == d].index # 先找到這個日期在交易日清單中位子，再往前三格確保轉倉日都是交易日
        s = s - 3
        switch = trading_day.iloc[s].iloc[0]['交易日']
        switch_date.append(switch)
    except:
        print('settlement_date', d, 'is not in the trading_day')
```

發現上述有些理論上的結算日並沒有交易，我手動上網查實際的資料，結果如下:  

1999/2/17 春節連假 改為1999/2/22結算  
2004/1/21 春節連假 改為2004/1/27結算  
2007/2/21 春節連假 改為2007/2/26結算  
2010/02/17 補假    改為2010/2/22結算    
2010/06/16 端午節休市 改為2010/6/17結算   
2013/08/21 因颱風休市 改為2013/8/22結算  
2015/2/18  適逢春假時期 改為2015/2/24結算    


以下重新調整結算日，從計算結算日的函數中增加例外的日期

```python
def find_third_wen(y, m):
    if y == 1999 and m == 2:
        d = 22
    elif y == 2004 and m == 1:
        d = 27
    elif y == 2007 and m == 2:
        d = 26
    elif y == 2010 and m == 2:
        d = 22
    elif y == 2010 and m == 6:
        d = 17
    elif y == 2013 and m == 8:
        d = 22
    elif y == 2015 and m == 2:
        d = 24
    elif datetime.date(y,m,1).weekday() < 2: #禮拜三還沒過
        d = 15 + (2 - datetime.date(y,m,1).weekday())
    elif datetime.date(y,m,1).weekday() == 2:#1號就是禮拜三
        d = 15
    else:
        d = (7 - datetime.date(y,m,1).weekday()) + 17#禮拜三過了
    date = datetime.datetime(y, m, d)
    return date
```

```python
year = [i for i in range(1998, 2023)]
month = [i for i in range(1, 13)]
#製作結算日清單
settlement_date = []
for y in year:
    for m in month:
        if y == 1998 and m in  [1, 2, 3, 4, 5, 6, 7, 8]:
            continue
        elif y == 2022 and m in [9, 10, 11, 12]:
            continue
        else:
            settlement = find_third_wen(y, m)
            settlement = str(settlement).split(' ')[0]
            settlement_date.append(settlement)
```

確定好結算日期以後，再跑一次找出轉倉日的程式碼

```python
switch_date = []
for d in settlement_date:
    try:
        s = trading_day[trading_day['交易日'] == d].index
        s = s - 3
        switch = trading_day.iloc[s].iloc[0]['交易日']
        switch_date.append(switch)
    except:
        print('settlement_date', d, 'is not in the trading_day')
```

```python
switch_date
```

確定結算日、轉倉日都是交易日之後，就可以依照trading_day的交易日清單，一個一個比對是否為結算日、轉倉日，然後再依據不同的情況抓出近月、次月的收盤價並計算轉倉利益  
以下這段程式包含幾個重點:  

1.非轉倉日時的轉倉利益皆為0  
2.遇到轉倉日時，當天的收盤價需改成次月期貨的收盤價  
3.轉倉日比結算日提前三個交易日，所以在這三個交易日中我們都要抓次月的開盤、收盤價(因為我們轉倉了但是期交所當月的合約還沒到期)，也就是底下程式碼 i = 1 的情況   
4.在當月的合約到期過後，代表轉倉、結算都已經完畢，此時要回到正常情況(i = 0)，也就是抓當月的開盤、收盤價

```python
final_data = []
last_close = 0 #用來暫時儲存前一日收盤價，計算轉倉利益
for i in range(len(trading_day)):
    date = trading_day.iloc[i]['交易日']
    if i == 1:#若我們已經先自行轉倉但期交所還沒結算當月期貨，則必須抓每個交易日的第二筆期貨價格資料
        df = day_data[day_data['日期'] == date]
        switch_gain = 0
        list = [df.iloc[1]['日期'], df.iloc[1]['開盤價'], df.iloc[1]['收盤價'], switch_gain]
        final_data.append(list)
        last_close = df.iloc[1]['收盤價']
        #如果剛好該日是結算日，則明天開始就要改回抓那天的第一筆期貨資料了，因為結算後前一天的次月就會變成當月
        if date in settlement_date:
            i = 0
        else:
            continue

    else:
        if not date in switch_date:#若不是轉倉日，則抓此交易日當月的期貨價格
            df = day_data[day_data['日期'] == date]
            switch_gain = 0
            list = [df.iloc[0]['日期'], df.iloc[0]['開盤價'], df.iloc[0]['收盤價'], switch_gain]
            final_data.append(list)
            last_close = df.iloc[0]['收盤價']
        #若遇到轉倉日，則收盤時需抓當日次月的收盤價，且往後三個交易日都需抓該日的第二筆可供交易的期貨價格(因為已經轉到次月的了)
        else:
            df = day_data[day_data['日期'] == date]
            switch_gain = df.iloc[1]['收盤價'] - last_close#以買方角度，計算今天收盤比前一天收盤多多少轉倉利益
            list = [df.iloc[0]['日期'], df.iloc[0]['開盤價'], df.iloc[1]['收盤價'], switch_gain]
            final_data.append(list)
            last_close = df.iloc[1]['收盤價']
            i = 1
```

```python
final_df = pd.DataFrame(final_data, columns = ['日期', '開盤價', '收盤價', '轉倉利益'])
```
